```python
# Leemos el csv
airlines = pd.read_csv('/Users/lucasgomeztorres/Downloads/airlines.csv',delimiter=',')
airports = pd.read_csv('/Users/lucasgomeztorres/Downloads/airports.csv',delimiter=',')

# Unir los datasets utilizando las iniciales de aeropuertos y aerolíneas
flights_merged = pd.merge(flights, airlines, left_on='Identificador de la Aerolínea', right_on='IATA_CODE', how='left')
flights_merged = pd.merge(flights_merged, airports, left_on='Aeropuerto de Origen', right_on='IATA_CODE', how='left')
flights_merged = pd.merge(flights_merged, airports, left_on='Aeropuerto de Destino', right_on='IATA_CODE', how='left')
```

```python
new_names = {'AIRLINE': 'Nombre de la aerolínea',
            'AIRPORT_x': 'Aeropuerto Origen',
            'CITY_x': 'Ciudad Origen',
             'STATE_x': 'Estado Origen',
             'LATITUDE_x': 'Latitud Origen',
             'LONGITUDE_x': 'Longitud Origen',
            'AIRPORT_y': 'Aeropuerto Destino',
            'CITY_y': 'Ciudad Destino',
             'STATE_y': 'Estado Destino',
             'LATITUDE_y': 'Latitud Destino',
             'LONGITUDE_y': 'Longitud Destino',
            
            }

# Renombramos las columnas 
flights_merged = flights_merged.rename(columns=new_names)

# Imprimir nombre columnas
print(flights_merged.columns.values)
```

```python
# Borramos las columnas innecesarias
flights_merged = flights_merged.drop("IATA_CODE_x", axis=1)
flights_merged = flights_merged.drop("IATA_CODE_y", axis=1)
flights_merged = flights_merged.drop("COUNTRY_x", axis=1)
flights_merged = flights_merged.drop("IATA_CODE", axis=1)
flights_merged = flights_merged.drop("Aeropuerto de Origen", axis=1)
flights_merged = flights_merged.drop("Aeropuerto de Destino", axis=1)
flights_merged = flights_merged.drop("COUNTRY_y", axis=1)
flights_merged = flights_merged.drop("Identificador de la Aerolínea", axis=1)
flights_merged = flights_merged.drop("Tiempo de Rodaje de Despegue", axis=1)
flights_merged = flights_merged.drop("Tiempo de Despegue de las ruedas", axis=1)
flights_merged = flights_merged.drop("Tiempo de vuelo en el aire", axis=1)
flights_merged = flights_merged.drop("Tiempo de Aterrizaje de las ruedas", axis=1)
flights_merged = flights_merged.drop("Tiempo Rodaje Aterrizaje", axis=1)
```

```python
# Crear campo tipo de fecha
from datetime import datetime
flights_merged['Fecha'] = pd.to_datetime(flights_merged['Día'].astype(str) + '-' + flights_merged['Mes'].astype(str) + '-' + 
                             flights_merged['Año'].astype(str), format='%d-%m-%Y')

print(flights_merged['Fecha'])
```

```python
import numpy as np

# Define el tamaño máximo de muestra para vuelos cancelados
tamano_max_cancelados = 500

# Crea una lista para almacenar las muestras balanceadas
muestras_balanceadas = []

# Itera sobre cada aerolínea
for aerolinea in flights_merged['Nombre de la aerolínea'].unique():
    # Filtra los vuelos de la aerolínea actual
    vuelos_aerolinea = flights_merged[flights_merged['Nombre de la aerolínea'] == aerolinea]
    
    # Separa los vuelos cancelados y no cancelados
    vuelos_cancelados = vuelos_aerolinea[vuelos_aerolinea['Cancelado'] == 1]
    vuelos_no_cancelados = vuelos_aerolinea[vuelos_aerolinea['Cancelado'] == 0]
    
    # Calcula el tamaño de muestra para vuelos cancelados
    tamano_muestra_cancelados = min(tamano_max_cancelados, len(vuelos_cancelados))
    
    # Calcula el tamaño de muestra para vuelos no cancelados
    tamano_muestra_no_cancelados = min(tamano_muestra_cancelados, len(vuelos_no_cancelados))
    
    # Realiza la selección aleatoria de los vuelos cancelados y no cancelados
    muestra_cancelados = vuelos_cancelados.sample(n=tamano_muestra_cancelados, random_state=42)
    muestra_no_cancelados = vuelos_no_cancelados.sample(n=tamano_muestra_no_cancelados, random_state=42)
    
    # Combina las muestras en una sola muestra balanceada para la aerolínea actual
    muestra_balanceada = pd.concat([muestra_cancelados, muestra_no_cancelados])
    
    # Agrega la muestra balanceada a la lista
    muestras_balanceadas.append(muestra_balanceada)

# Combina todas las muestras balanceadas en un único DataFrame
vuelos_balanceados = pd.concat(muestras_balanceadas)
```
